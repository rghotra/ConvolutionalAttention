```python
baselines = ['CNN_LSTM']
categories = ['CNN_LSTM']
variants = [[('relu', 4), ('relu', 25), ('exponential', 4), ('exponential', 25)]]
names = [['pool4_relu', 'pool25_relu', 'pool4_exp', 'pool25_exp']]
trials = 5
```

```python
total = 0
for a in range(len(baselines)):
    for i in range(len(categories)):
        for j in range(len(variants[i])):
            for k in range(1, trials+1):
                # model_num = a*len(categories)*len(variants[i])*trials + i*len(variants[i])*trials + j*trials + k
                # print('\n' + str(model_num/1.6) + '% completed')
                print('\n' + baselines[a] + ' ' + categories[i] + ' model-' + str(names[i][j]) + ' trial-' + str(k))

                # num_heads = b_heads if categories[i] != 'heads' else variants[i][j]
                # key_size = b_key if categories[i] != 'key_sizes' else variants[i][j]
                # pool_size = 4 if a % 2 == 0 else 25
                # activation = 'relu' if a < 2 else 'exponential'

                # model = models.CNN_ATT(activation=activation, pool_size=pool_size, heads=num_heads, key_size=key_size)
                
                activation = variants[i][j][0]
                pool_size = variants[i][j][1]
                
                model = models.CNN_LSTM(activation=activation, pool_size=pool_size)
                
                pipeline.execute_pipeline(baselines[a], categories[i], names[i][j], k, model, 100)
                
                # pipeline.execute_pipeline(baselines[a], categories[i], variants[i][j], k, model, 100)
                
                total += 1
```
