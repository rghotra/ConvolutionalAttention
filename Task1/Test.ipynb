```python
import utils, models
import tensorflow as tf
from six.moves import cPickle
```

```python
a, b, c, d, x_test, y_test = utils.get_synthetic_dataset()
```

```python
for i in range(1, 6):
    model = models.CNN_LSTM_ATT(activation='exponential')
    model.load_weights(f'CNN_LSTM_ATT/models/CNN_LSTM_ATT/model-pool4_exp/trial-{i}/weights')
    
    auroc = tf.keras.metrics.AUC(curve='ROC', name='auroc')
    aupr = tf.keras.metrics.AUC(curve='PR', name='aupr')
    model.compile(tf.keras.optimizers.Adam(0.0005), loss='binary_crossentropy', metrics=[auroc, aupr])
    
    model.evaluate(x_test, y_test)
```

```python
for i in range(1, 6):
    with open(f'CNN_LSTM_ATT/history/CNN_LSTM_ATT/model-pool4_exp/trial-{i}.pickle', 'rb') as file:
        history = cPickle.load(file)
    print(history['val_aupr'][-1])
```
